## 2. Accessing Corpora

```python
# Import NLTK and Load Corpora:

import nltk
from nltk.corpus import gutenberg
```

```python
# Explore Available Corpora:
print(nltk.corpus.gutenberg.fileids())
```

```python
# Access Text Data:
text = gutenberg.raw('austen-emma.txt')
```

```python
# Viewing Sample Data:
print(text[:500])  # Prints the first 500 characters
```

## 3. Manipulating Corpora


```python
# Tokenization:

from nltk.tokenize import word_tokenize, sent_tokenize

words = word_tokenize(text)
sentences = sent_tokenize(text)
```

```python
# Frequency Distribution:
from nltk.probability import FreqDist

fdist = FreqDist(words)
print(fdist.most_common(10))  # Print the 10 most common words
```

```python
# Concordance
from nltk.text import Text

nltk_text = Text(words)
nltk_text.concordance('Emma')
```

```python
# Collocations

from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures

bigram_finder = BigramCollocationFinder.from_words(words)
bigrams = bigram_finder.nbest(BigramAssocMeasures.likelihood_ratio, 10)
print(bigrams)
```
