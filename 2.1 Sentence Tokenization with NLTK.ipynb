# 1. Sentence Tokenization

- process of splitting block of text into sentences.
- Applications:
  - semantic analysis
  - text summarization
  - question answering

```python
# import libraries

import nltk
nltk.download('punkt')  # 'punkt' is the tokenizer model used by NLTK for sentence tokenization
```

```python
# data
text = "Hello! How are you today? I hope you're doing well. Have a great day ahead."
```

```python
# Perform Sentence Tokenization:

from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text)
print(sentences)

```

```python
# output

['Hello!', 'How are you today?', 'I hope you\'re doing well.', 'Have a great day ahead.']
```
