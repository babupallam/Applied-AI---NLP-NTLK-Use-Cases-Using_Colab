```python
import nltk
from nltk.chat.util import Chat, reflections
```

```python
pairs = [
    (r'hi|hello|hey', 'Hello! How can I help you today?'),
    (r'what is your name?', 'I am a chatbot created to assist you.'),
    (r'how are you?', 'I am just a program, but I am functioning as expected!'),
    (r'quit', 'Bye! Have a nice day.')
]
```

```python
#creating chatbot

chatbot = Chat(pairs, reflections)
```

```python
#implement chatbot loop

def chat():
    print("Hello! I'm your chatbot. Type 'quit' to exit.")
    while True:
        user_input = input("You: ")
        response = chatbot.respond(user_input)
        print("Chatbot:", response)
        if user_input.lower() == 'quit':
            break

if __name__ == "__main__":
    chat()
```

```python
import nltk
from nltk.chat.util import Chat, reflections
```

```python
pairs = [
    [
        r"my name is (.*)",
        ["Hello %1, How can I help you today?",]
    ],
    [
        r"hi|hello|hey",
        ["Hello! How can I assist you?", "Hey there! What can I do for you?", "Hi! How may I help you?"]
    ],
    [
        r"what is your name?",
        ["I am a bot created by OpenAI. You can call me ChatGPT!",]
    ],
    [
        r"how are you?",
        ["I'm just a bot, but I'm here to help! How are you?",]
    ],
    [
        r"sorry (.*)",
        ["No problem at all!", "It's okay!", "No need to apologize."]
    ],
    [
        r"i'm (.*) doing good",
        ["Great to hear! How can I assist you?",]
    ],
    [
        r"what can you do?",
        ["I can chat with you, answer questions, and help with basic tasks. What would you like to know?",]
    ],
    [
        r"quit",
        ["Goodbye! It was nice talking to you.", "See you later!"]
    ],
    [
        r"(.*)",
        ["I'm not sure I understand. Can you please rephrase?",]
    ]
]
```

```python
chatbot = Chat(pairs, reflections)
```

```python
# interaction
def start_chat():
    print("Hi! I am a simple chatbot. Type 'quit' to exit.")
    chatbot.converse()

if __name__ == "__main__":
    start_chat()
```
