# 6. Parsing and Syntactic Analysis with NLTK

Parsing is a crucial task in natural language processing (NLP) that involves analyzing the grammatical structure of a sentence. It identifies the roles of words and how they relate to each other, enabling more complex language understanding tasks like sentence generation, translation, and question answering.

```python
import nltk
from nltk import CFG
from nltk import ChartParser
```

```python
# Define a Context-Free Grammar (CFG)

grammar = CFG.fromstring("""
  S -> NP VP
  NP -> DT NN | DT JJ NN | PRP
  VP -> VBZ NP | VBZ NP PP
  PP -> IN NP
  DT -> 'the' | 'a'
  NN -> 'dog' | 'park'
  JJ -> 'big'
  VBZ -> 'chases' | 'sees'
  IN -> 'in'
  PRP -> 'he'
""")
```

```python
# Create the Parser

parser = ChartParser(grammar)
```

```python
# Parse a Sentence
sentence = "the big dog chases a dog".split()
for tree in parser.parse(sentence):
    print(tree)
```
